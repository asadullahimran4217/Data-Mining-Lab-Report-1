```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid", context="talk")

df = pd.read_csv("https://raw.githubusercontent.com/Ataullha/CSE-412_Machine-Learning-Lab/refs/heads/main/income.csv")

print(df.head())
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, linewidths=0.5)
plt.title("Correlation Heatmap of Income Dataset")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(data=df, y="Income")
plt.title("Boxplot of Income")
plt.tight_layout()
plt.savefig("figures/boxplot.pdf")
plt.show()
```
